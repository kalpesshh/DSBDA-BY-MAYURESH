```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("covid_vaccine_statewise.csv")
df
```

```python
df.head()
```

```python
first_dose_by_state = df.groupby('State')['First Dose Administered'].sum().reset_index()
```

```python
second_dose_by_state = df.groupby('State')['Second Dose Administered'].sum().reset_index()
```

```python
males_by_state = df.groupby('State')['Male(Individuals Vaccinated)'].sum().reset_index()
```

```python
females_by_state = df.groupby('State')['Female(Individuals Vaccinated)'].sum().reset_index()
```

```python
print("Number of individuals vaccinated for the first dose by state:")
print(first_dose_by_state)
```

```python
print("Number of individuals vaccinated for the second dose by state:")
print(second_dose_by_state)
```

```python
print("Number of males vaccinated by state:")
print(males_by_state)
```

```python
print("Number of females vaccinated by state:")
print(females_by_state)
```

```python
vaccinated_by_state = df.groupby('State')['Total Individuals Vaccinated'].sum().reset_index()
```

```python
total_vaccinated = vaccinated_by_state['Total Individuals Vaccinated'].sum()
```

```python
vaccinated_by_state['Total Individuals Vaccinated'] = vaccinated_by_state['Total Individuals Vaccinated'].apply(lambda x: int(x))
```

```python
vaccinated_by_state = vaccinated_by_state.sort_values('Total Individuals Vaccinated', ascending=False)
```

```python
print(vaccinated_by_state)
```

```python
plt.figure(figsize=(12,6))
plt.bar(vaccinated_by_state['State'], vaccinated_by_state['Total Individuals Vaccinated']/total_vaccinated)
plt.xticks(rotation=90)
plt.xlabel('State')
plt.ylabel('Proportion of Vaccinated Individuals')
plt.title('Proportion of Vaccinated Individuals by State')
plt.show()
```
